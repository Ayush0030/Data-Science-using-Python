```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
Y=stats.norm()
X=np.linspace(-5,5,100)

fig,axes=plt.subplots(3,1,sharex=True)



axes[0].plot(X,Y.pdf(X))
#axes[0].plot(np.arange(10),np.arange(20,30))


axes[1].plot(X,Y.cdf(X))

axes[2].hist(Y.rvs(size=1000),bins=50)
```

```python
#image processing

import scipy.ndimage as ndimage

#raccoon

from scipy import misc
raccoon= misc.face()

plt.imshow(raccoon)
plt.show()
```

```python
filtered= ndimage.uniform_filter(raccoon,size=40)
plt.imshow(filtered)
```

```python
face=misc.face()
flip_up=np.flipud(face)
plt.imshow(flip_up)
```
