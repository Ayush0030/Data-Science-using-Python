```python
#optimize
from scipy import optimize
```

```python
#scipy.optimise 
```

```python
import numpy as np
def f(x):
    return -np.sin(x)

#function fx having negative sinx
#x axis
x=np.linspace(0,5,100)

#point
point=3

optimized=optimize.fmin(f,point)

import matplotlib.pyplot as plt
plt.plot(x,f(x))
plt.scatter(point,f(point))
plt.scatter(optimized,f(optimized))
plt.legend(['Function -sin(x)', 'Point','Optimized'])
```

```python
import numpy as np
def f(x):
    return x**2 + x*2 +1


#function fx having negative sinx
#x axis
x=np.linspace(0,2,10)

#point
point=3

optimized=optimize.fmin(f,point)

import matplotlib.pyplot as plt
plt.plot(x,f(x))
plt.scatter(point,f(point))
plt.scatter(optimized,f(optimized))
```

```python
from scipy import stats
```

```python
#statiscal modelling --> StatsModels
```

```python
X=stats.poisson(3.5)
n=np.arange(0,15)

fig,axes=plt.subplots(3,1,sharex=True)
#probabilty mass function
axes[0].step(n,X.pmf(n))
#commu;ativre distribution fuinction
axes[1].step(n,X.cdf(n))
#100 random values histogram
axes[2].hist(X.rvs(size=1000))
```
