```python
x, y=np.random.multivariate_normal([0,0],[[1,-0.5],[-0.5,1]],size=300).T
sample_cmap=sns.cubehelix_palette(as_cmap=True)
sns.kdeplot(x,y, cmap=sample_cmap, shade=True)
#Covariance matrix
```
