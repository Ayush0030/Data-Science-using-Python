```python
import seaborn as sns
import matplotlib.pyplot as plt
from pylab import *
import numpy as np
import pandas as pd
```

```python
iris=sns.load_dataset("iris")
iris
```

```python
setosa=iris.loc[iris.species=="setosa"]
virginica=iris.loc[iris.species=="virginica"]
```

```python
sns.kdeplot(setosa.sepal_length, color="r", shade=True)
```

```python
sns.histplot(iris.sepal_length, kde=True)
```

```python
sns.histplot(iris.sepal_length, kde=True, color="r")
sns.histplot(iris.sepal_width, kde=True, color="k")
```

```python
sns.kdeplot(data=iris, x="sepal_length", y="sepal_width")
```

```python
sns.kdeplot(data=iris, x="sepal_length", y="sepal_width",  shade=True)
```

```python
# "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" 
sns.jointplot(data=iris,x="sepal_length",y="sepal_width", kind="hex")
```
