```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import Delaunay
```

```python
#The Delaunay Triangulation is a sub division
#of set of points into a non overlapping set of triangles 
#(such that no points lie inside the triangulation)
##here we tend to avoid small acute angles (angles that lie between 0 to 90) 
```

```python
MYPOINT=np.array([[0,3],[1,1.1],[1,4],[1,2]])
MYPOINT
TRIANGLE=Delaunay(MYPOINT)
TRIANGLE
plt.triplot(MYPOINT[:,0],MYPOINT[:,1])
plt.plot(MYPOINT[:,0],MYPOINT[:,1],'o')
print(MYPOINT)
print(MYPOINT[:,0])
```

```python
tammana_points=np.array([[0,4],[2,1.1],[1,3],[1,2]])
TRIANGLE_tammana=Delaunay(tammana_points)
plt.triplot(tammana_points[:,0],tammana_points[:,1])
plt.plot(tammana_points[:,0],tammana_points[:,1],'o')
```

```python
tammana_points=np.array([[0,0],[0,1.1],[1,0],[1,1]])
TRIANGLE_tammana=Delaunay(tammana_points)
plt.triplot(tammana_points[:,0],tammana_points[:,1])
plt.plot(tammana_points[:,0],tammana_points[:,1],'o')
```
