```python
from scipy import interpolate
import numpy as np
import matplotlib.pyplot as plt
```

```python
#scatter points
x_array=np.arange(0,10,0.5)
y_array=np.sin(x_array)


ff= interpolate.interp1d(x_array,y_array)
#interpolation function
x_array_i=np.arange(0,10,3)
#0 3 6 9
y_array_i=ff(x_array_i)
#interpolation points


plt.scatter(x_array,y_array)
plt.plot(x_array_i,y_array_i)

plt.legend(['Interpolation','Data Points'])
```

```python
x=np.arange(0,10,0.5)
y=np.sin(x)
function_new1= interpolate.interp1d(x,y,kind="linear")
function_new2= interpolate.interp1d(x,y,kind="cubic")
x_new=np.linspace(0,4,30)

plt.plot(x,y,'o',x_new,function_new1(x_new),'_',x_new,function_new2(x_new),'--')
plt.legend(['data','linear','cubic','nearest'],loc="best")
```

```python
frequency_sample=100
Time=np.linspace(0,1,frequency_sample*2,endpoint=False)
f1,f2,f3=1,5,20
A1,A2,A3=3,2,1
x1=A1*np.sin(f1*2*np.pi*Time)
x2=A2*np.sin(f2*2*np.pi*Time)
x3=A3*np.sin(f3*2*np.pi*Time)

x=x1+x2+x3
#one graph would be using x
#another graph would be having x1,x2,x3

plt.subplot(2,1,1)
plt.plot(Time,x1,Time,x2,Time,x3)

plt.subplot(2,1,2)
plt.plot(Time,x)

plt.xlabel('Time')
plt.ylabel('Amplitude')
```
