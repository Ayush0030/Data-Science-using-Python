```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
slices=[60,40]
plt.pie(slices)
```

```python
plt.pie(slices)
plt.grid(b=True)
plt.show()
```

```python
labels=['Sixty','Forty']
plt.pie(slices,labels=labels)
plt.show
```

```python
plt.pie(slices,labels=labels,shadow=True,wedgeprops={'edgecolor':'white'})
```

```python
plt.pie(slices,labels=labels,wedgeprops={'edgecolor':'white'})
```

```python
plt.pie(slices,labels=labels,wedgeprops={'edgecolor':'red','linewidth':9})
```

```python
colors=['yellow','green']
plt.pie(slices, labels=labels, colors=colors, wedgeprops={'edgecolor':'red','linewidth':3})
```

```python
vinayak_slices=[80,120,30,25,99,119]
my_labels=["Match 1","Match 2","Match 3","Match 4","Match 5","Match 6"]
My_colors=['#80ADD7','#0ABDA0','#EBF3EA','#D4DCA9','#BF9D7A','pink']
Vinayak_explode=[0,0,0,0.9,0,0]
plt.pie(vinayak_slices,labels=my_labels,explode=Vinayak_explode,colors=My_colors,wedgeprops={'edgecolor':'black'})
```
