```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.linspace(10)
```

```python
x=np.random.rand(100)
print(x)
y=np.random.rand(100)
print(y)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y,color='#73ff00')
```

```python
plt.scatter(x,y,color='#73ff00',s=234)
```

```python
plt.scatter(x,y,color='#73ff00',alpha=0.2,s=234)
```

```python
plt.scatter(x,y,color='#73ff00')
```

```python
plt.scatter(x,y,color='#73ff00',edgecolor='black')
```

```python

plt.scatter(x,y,c=np.arange(100),s=234,cmap='gist_ncar')
```

```python
plt.scatter(x,y,c=np.arange(100),s=234,alpha=0.5,edgecolor='black')
```

```python
plt.scatter(x,y,c=np.random.rand(100),s=300,edgecolor='black',marker=(5,1))
```

```python
x_new=np.random.randn(100)
y_new=np.random.randn(100)
my_colors=np.random.randn(100)
my_sizes=1000*np.random.rand(100)

plt.scatter(x_new,y_new,c=my_colors,s=my_sizes,alpha=0.5,cmap='gist_ncar')
plt.colorbar()
```
