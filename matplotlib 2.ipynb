```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([0,1,2,3,4,5])
y=x**2
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
#left bottom width height
axes.plot(x,y)
axes.set_xlabel("X label")
axes.set_ylabel('Y Label')
axes.set_title('My Graph')
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
#left bottom width height
axes.plot(x,y,label="y=x^2")
axes.set_xlabel("X label")
axes.set_ylabel('Y Label')
axes.set_title('My Graph')
axes.legend()
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
#left bottom width height
axes.plot(x,y,label="y=x^2")
axes.set_xlabel("X label")
axes.set_ylabel('Y Label')
axes.set_title('My Graph')
axes.legend(loc=4)
```

```python
fig, axes=plt.subplots()
axes.plot(x,y,label="y=x^2")
axes.plot(y,x,label="opposite to y=x^2")
axes.legend()
axes.set_xlabel('X Axis Values')
axes.set_ylabel('Y Axis Values')
```

```python
matplotlib.rcParams.update({'font.size':25,'font.family':'serif'})
fig, axes=plt.subplots()
axes.plot(x,y,label="y=x^2")
axes.plot(y,x,label="opposite to y=x^2")
axes.legend()
axes.set_xlabel('X Axis Values')
axes.set_ylabel('Y Axis Values')
```

```python

matplotlib.rcParams.update({'font.size':25,'font.family':'serif'})


fig, axes=plt.subplots(figsize=(40,20))
axes.plot(x,y,label="y=x^2")
axes.plot(y,x,label="opposite to y=x^2")
axes.legend()
axes.set_xlabel('X Axis Values')
axes.set_ylabel('Y Axis Values')
```

```python
fig, axes=plt.subplots()
axes.plot(x,y,'--',label="y=x^2")
axes.plot(x,x**3,'.-',label="y=x^3")
axes.legend()
axes.set_xlabel('X Axis Values')
axes.set_ylabel('Y Axis Values')
```

```python
fig, axes=plt.subplots()
axes.plot(x,y,'r--',label="y=x^2")
axes.plot(x,x**3,label="y=x^3")
axes.legend()
axes.set_xlabel('X Axis Values')
axes.set_ylabel('Y Axis Values')
```

```python
fig, axes=plt.subplots()
axes.plot(x,y,color="#00fff7",label="y=x^2")
axes.plot(x,x**3,label="y=x^3")
axes.legend()
axes.set_xlabel('X Axis Values')
axes.set_ylabel('Y Axis Values')
```

```python
fig,axes=plt.subplots()
axes.plot(x,x+2,label="x+2")
axes.plot(x,x+4,label="x+4")
axes.plot(x,x+6,label="x+6")
axes.plot(x,x+8,color="pink",label="x+8")
axes.legend()
```

```python
fig,axes=plt.subplots(figsize=(40,20))

axes.plot(x,x+1,color="black",linewidth=0.3)
axes.plot(x,x+2,color="black",linewidth=5)
axes.plot(x,x+3,color="black",linewidth=15)
axes.plot(x,x+4,color="black",linewidth=35)


```

```python
fig,axes=plt.subplots(figsize=(40,20))

axes.plot(x,x+1,color="black",linewidth=15,linestyle='-.',label="x+1")
axes.plot(x,x+2,color="blue",linewidth=15,linestyle='--',label="x+2")
axes.plot(x,x+3,color="red",linewidth=15,linestyle=':',label="x+3")
axes.plot(x,x+4,color="green",linewidth=15,linestyle='-',label="x+4")
axes.legend()
```

```python
fig,axes=plt.subplots(figsize=(40,20))

axes.plot(x,x+1,color="black",linewidth=15,linestyle='-.',label="x+1",marker='+',markersize=155)
axes.plot(x,x+2,color="blue",linewidth=15,linestyle='--',label="x+2",marker='o',markersize=45)
axes.plot(x,x+3,color="red",linewidth=15,linestyle=':',label="x+3",marker='*',markersize=155)
axes.plot(x,x+4,color="green",linewidth=15,linestyle='-',label="x+4",marker='1',markersize=155)
axes.legend()
```
