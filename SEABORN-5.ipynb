```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style="darkgrid")
```

```python
dataframe=sns.load_dataset('iris')
dataframe.head()
```

```python
sns.violinplot(x=dataframe["species"],y=dataframe["sepal_length"])
```

```python
sns.violinplot(x=dataframe["species"],y=dataframe["sepal_length"],width=2)
```

```python
sns.violinplot(x=dataframe["species"],y=dataframe["sepal_length"],palette="icefire")
```

```python
sns.violinplot(x=dataframe["species"],y=dataframe["sepal_length"],palette="Blues")
```

```python
sns.violinplot(y=dataframe["species"],x=dataframe["sepal_length"])
```
